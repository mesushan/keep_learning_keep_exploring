#### Approach 3: Embed an image by Base64 Encode-Decode
- Base64 encoding algorithm is a binary to text encoding algorithm to convert data to plain text.
- This algorithm can encode/decode audio or video files too.
- First we need to encode the image using [Base64-Image](https://www.base64-image.de/) and then decode it in the notebook using the Python standard base64 library.
- Important thing to notice -> Remove `data:image/png;base64,` form the encoded data which will be in the beginning of encoded image
- Using this method, you don't have to rely on the external resource as they can be removed at any time.

```python
from base64 import b64decode

encoded_data = ""
display.Image(b64decode(encoded_data))
```
